```python
import numpy as np
import random
import pandas as pd

def randomboard() :
    """Génère un plateau aléatoire 3x3 avec pions blancs (1), noirs (-1) et cases vides (0)."""
    board = np.zeros((3, 3), dtype=int)
    # print("board : \n", board)
    
    num_white = 3
    num_black = 3
    
    # Placer les pions blancs
    positions = random.sample(range(9), num_white)
    for pos in positions:
        board[pos // 3][pos % 3] = 1  # Blanc
    
    # print("white on board: \n", board)
    
    # Placer les pions noirs
    available_positions = [i for i in range(9) if board[i // 3][i % 3] == 0]
    positions = random.sample(available_positions, num_black)
    for pos in positions:
        board[pos // 3][pos % 3] = -1  # Noir
    
    print("black and white on board : \n", board)

    return board

#verification si les blancs gagnet ou perdent
def check_victory(board, player):
    """Vérifie si 'player' (1 pour Blancs, -1 pour Noirs) a gagné."""
    # Vérification des lignes et colonnes
    for i in range(3):
        if np.all(board[i, :] == player) or np.all(board[:, i] == player):
            return True
    
    # Vérification des diagonales
    if np.all(np.diag(board) == player) or np.all(np.diag(np.fliplr(board)) == player):
        return True
    
    return False

# Génération des positions gagnantes et perdantes
winning_positions = []
losing_positions = []

while len(winning_positions) < 500 or len(losing_positions) < 500:
    board = randomboard()
    
    if check_victory(board, 1) and len(winning_positions) < 500:
        winning_positions.append(board.flatten().tolist())
    elif check_victory(board, -1) and len(losing_positions) < 500:
        losing_positions.append(board.flatten().tolist())

# print("winning position : \n", winning_positions)
# print("losing position : \n", losing_positions)

# Sauvegarde des positions en fichiers CSV
df_win = pd.DataFrame(winning_positions)
df_loss = pd.DataFrame(losing_positions)

df_win.to_csv("fanorona_winning_positions.csv", index=False)
df_loss.to_csv("fanorona_losing_positions.csv", index=False)

print("500 positions gagnantes et 500 perdantes générées et sauvegardées.")

df_win = pd.read_csv("dataset_fanorona/fanorona_winning_positions.csv")
df_loose = pd.read_csv("dataset_fanorona/fanorona_losing_positions.csv")
print("Nombre de doublons :", df_win.duplicated().sum())
print("Nombre de doublons :", df_loose.duplicated().sum())

df_win.drop_duplicates(inplace=True)
df_loose.drop_duplicates(inplace=True)

# Save the cleaned DataFrames to new CSV files
df_win.to_csv("dataset_fanorona/fanorona_winning_positions_cleaned.csv", index=False)
df_loose.to_csv("dataset_fanorona/fanorona_losing_positions_cleaned.csv", index=False)

print("Fichiers nettoyés et sauvegardés avec succès.")
```
